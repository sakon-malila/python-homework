## 8.การสลับค่าตัวแปร

```python
a = 10
b = 20



print("a : {} , b: {}" .format(a, b))
# ผลที่คาดหวังคือ
# a : 20 , b: 10
```

```python
a = 10
b = 20

tmp = a    # เก็บค่าของ a ไว้ใน temp (temp = 10)
a = b      # เอาค่าของ b (20) มาใส่ใน a  → ตอนนี้ a = 20
b = tmp    # เอาค่าของ temp (10) มาใส่ใน b → ตอนนี้ b = 10

print("a : {} , b: {}" .format(a, b))
# ผลที่คาดหวังคือ
# a : 20 , b: 10
```

```python
a = 10
b = 20

# parallel assigment 
a, b = b, a

print("a : {} , b: {}" .format(a, b))
# ผลที่คาดหวังคือ
# a : 20 , b: 10
```
