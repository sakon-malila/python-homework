### ‡∏Ç‡πâ‡∏≠ 1: ‡∏Å‡∏≤‡∏£‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£‡∏™‡∏ï‡∏£‡∏¥‡∏á (String Manipulation)


3. ‡πÅ‡∏¢‡∏Å‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡∏≠‡∏≠‡∏Å‡∏°‡∏≤ ‡πÅ‡∏•‡∏∞‡πÅ‡∏õ‡∏•‡∏á‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô integer
4. ‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå‡πÉ‡∏ô‡∏£‡∏π‡∏õ‡πÅ‡∏ö‡∏ö‡∏ó‡∏µ‡πà‡∏≠‡πà‡∏≤‡∏ô‡∏á‡πà‡∏≤‡∏¢

````python
student_info = "‡∏£‡∏´‡∏±‡∏™:6601234|‡∏ä‡∏∑‡πà‡∏≠:‡∏™‡∏°‡∏ä‡∏≤‡∏¢ ‡πÉ‡∏à‡∏î‡∏µ|‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô:85"

# ‡∏û‡∏¥‡∏°‡∏û‡πå‡πÇ‡∏Ñ‡πâ‡∏î‡∏Ñ‡∏≥‡∏ï‡∏≠‡∏ö‡∏ó‡∏µ‡πà‡∏ô‡∏µ‡πà
part = student_info.split("|")
stu_id = part[0].split(":")[1]
name = part[1].split(":")[1]
score = int(part[2].split(":")[1])
print(f"‡∏£‡∏´‡∏±‡∏™‡∏ô‡∏±‡∏Å‡∏®‡∏∂‡∏Å‡∏©‡∏≤: {stu_id}")
print(f"‡∏ä‡∏∑‡πà‡∏≠: {name}")
print(f"‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô: {score}")
````

5. ‡πÅ‡∏™‡∏î‡∏á‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î
6. ‡πÅ‡∏™‡∏î‡∏á‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤ 3 ‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡πÅ‡∏£‡∏Å

**Expected Output:**
```
‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤: ['‡∏ô‡∏°', '‡∏Ç‡∏ô‡∏°‡∏õ‡∏±‡∏á', '‡∏ä‡∏≤', '‡∏ô‡πâ‡∏≥‡∏ï‡∏≤‡∏•', '‡πÄ‡∏ô‡∏¢', '‡πÅ‡∏¢‡∏°']
‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î: 6 ‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£
‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤ 3 ‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡πÅ‡∏£‡∏Å: ['‡∏ô‡∏°', '‡∏Ç‡∏ô‡∏°‡∏õ‡∏±‡∏á', '‡∏ä‡∏≤']
```

````python
products = ["‡∏ô‡∏°", "‡∏Ç‡∏ô‡∏°‡∏õ‡∏±‡∏á", "‡πÑ‡∏Ç‡πà", "‡∏ô‡πâ‡∏≥‡∏ï‡∏≤‡∏•"]

# ‡∏û‡∏¥‡∏°‡∏û‡πå‡πÇ‡∏Ñ‡πâ‡∏î‡∏Ñ‡∏≥‡∏ï‡∏≠‡∏ö‡∏ó‡∏µ‡πà‡∏ô‡∏µ‡πà
products.append("‡πÄ‡∏ô‡∏¢")
products.append("‡πÅ‡∏¢‡∏°")
products.insert(2,"‡∏ä‡∏≤")
products.remove("‡πÑ‡∏Ç‡πà")
print(f"‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤: {products}")
print(f"‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î: {len(products)}")
print(f"‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤ 3 ‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡πÅ‡∏£‡∏Å: {products[0:3]}")
````

1. ‡∏™‡∏£‡πâ‡∏≤‡∏á dictionary ‡∏Ç‡∏≠‡∏á‡∏ô‡∏±‡∏Å‡∏®‡∏∂‡∏Å‡∏©‡∏≤ 3 ‡∏Ñ‡∏ô ‡πÇ‡∏î‡∏¢‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏Ñ‡∏ô‡∏°‡∏µ ‡∏£‡∏´‡∏±‡∏™, ‡∏ä‡∏∑‡πà‡∏≠, ‡πÅ‡∏•‡∏∞ ‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô
2. ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏ô‡∏±‡∏Å‡∏®‡∏∂‡∏Å‡∏©‡∏≤‡∏Ñ‡∏ô‡∏ó‡∏µ‡πà 4 ‡∏•‡∏á‡πÉ‡∏ô dictionary
3. ‡∏≠‡∏±‡∏õ‡πÄ‡∏î‡∏ï‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡∏Ç‡∏≠‡∏á‡∏ô‡∏±‡∏Å‡∏®‡∏∂‡∏Å‡∏©‡∏≤‡∏Ñ‡∏ô‡πÉ‡∏î‡∏Ñ‡∏ô‡∏´‡∏ô‡∏∂‡πà‡∏á
4. ‡πÅ‡∏™‡∏î‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ô‡∏±‡∏Å‡∏®‡∏∂‡∏Å‡∏©‡∏≤‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î (‡πÉ‡∏´‡πâ‡πÉ‡∏ä‡πâ loop)
5. ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢‡∏Ç‡∏≠‡∏á‡∏ô‡∏±‡∏Å‡∏®‡∏∂‡∏Å‡∏©‡∏≤‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î

**Expected Output:**
```

‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢: XX.XX

[‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏Ñ‡∏ô]
```

````python
# ‡∏û‡∏¥‡∏°‡∏û‡πå‡πÇ‡∏Ñ‡πâ‡∏î‡∏Ñ‡∏≥‡∏ï‡∏≠‡∏ö‡∏ó‡∏µ‡πà‡∏ô‡∏µ‡πà

students = {
    "6601001": {"name": "‡∏™‡∏°‡∏ä‡∏≤‡∏¢", "score": 78},
    "6601002": {"name": "‡∏™‡∏°‡∏´‡∏ç‡∏¥‡∏á", "score": 85},
    "6601003": {"name": "‡∏™‡∏°‡∏´‡∏ß‡∏±‡∏á", "score": 92}
}
students[6601004] = {
    "name": "‡∏™‡∏°‡πÅ‡∏•‡πâ‡∏ß",
    "score": 100
}
students["6601002"]["score"] = 86

print(f"‡∏ô‡∏±‡∏Å‡∏®‡∏∂‡∏Å‡∏©‡∏≤‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î: {len(students)}")
total_score = 0
for info in students.values():
    total_score += info["score"]
avg_score = total_score / len(students)
print(f"‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢ = {avg_score:.2f}")

print()
for student_id, info in students.items():
    print(f"‡∏£‡∏´‡∏±‡∏™: {student_id}")
    print(f"‡∏ä‡∏∑‡πà‡∏≠: {info['name']}")
    print(f"‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô: {info['score']}")
    print("-" * 20)
````

### ‡∏Ç‡πâ‡∏≠ 4: ‡∏ô‡∏¥‡∏û‡∏à‡∏ô‡πå‡∏ó‡∏≤‡∏á‡∏Ñ‡∏ì‡∏¥‡∏ï‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå (Mathematical Expressions)

‡πÉ‡∏´‡πâ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡πÇ‡∏õ‡∏£‡πÅ‡∏Å‡∏£‡∏°‡πÅ‡∏õ‡∏•‡∏á‡∏ô‡∏¥‡∏û‡∏à‡∏ô‡πå‡∏ó‡∏≤‡∏á‡∏Ñ‡∏ì‡∏¥‡∏ï‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå‡∏ï‡πà‡∏≠‡πÑ‡∏õ‡∏ô‡∏µ‡πâ‡πÄ‡∏õ‡πá‡∏ô‡πÇ‡∏Ñ‡πâ‡∏î Python ‡πÅ‡∏•‡∏∞‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå:

1. **‡∏û‡∏∑‡πâ‡∏ô‡∏ó‡∏µ‡πà‡∏ß‡∏á‡∏Å‡∏•‡∏°:** A = œÄr¬≤ ‡πÇ‡∏î‡∏¢‡πÉ‡∏´‡πâ r = 5
2. **‡∏ô‡∏¥‡∏û‡∏à‡∏ô‡πå‡πÄ‡∏®‡∏©‡∏™‡πà‡∏ß‡∏ô:** (15 + 25) √ó 3 √∑ 2‚Å¥
3. **‡∏£‡∏≤‡∏Å‡∏ó‡∏µ‡πà 3:** ‚àõ64 (cube root of 64)
4. **‡∏£‡∏∞‡∏¢‡∏∞‡∏ó‡∏≤‡∏á:** d = ‚àö((x‚ÇÇ-x‚ÇÅ)¬≤ + (y‚ÇÇ-y‚ÇÅ)¬≤) ‡πÇ‡∏î‡∏¢‡πÉ‡∏´‡πâ (x‚ÇÅ,y‚ÇÅ) = (3,4) ‡πÅ‡∏•‡∏∞ (x‚ÇÇ,y‚ÇÇ) = (7,1)

**‡∏´‡∏°‡∏≤‡∏¢‡πÄ‡∏´‡∏ï‡∏∏:** ‡πÉ‡∏ä‡πâ `math` module ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö pi ‡πÅ‡∏•‡∏∞ sqrt

**Expected Output:**
```
1. ‡∏û‡∏∑‡πâ‡∏ô‡∏ó‡∏µ‡πà‡∏ß‡∏á‡∏Å‡∏•‡∏° = 78.50
2. ‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå‡∏ô‡∏¥‡∏û‡∏à‡∏ô‡πå = 7.50
3. ‡∏£‡∏≤‡∏Å‡∏ó‡∏µ‡πà 3 ‡∏Ç‡∏≠‡∏á 64 = 4.00
4. ‡∏£‡∏∞‡∏¢‡∏∞‡∏ó‡∏≤‡∏á = 5.00
```

````python
import math
pi = math.pi
r = 5

A = pi * (r**2)
B = (15 + 25) * (3/(2**4))
C = 64 ** (1/3)
D = (((7-3)**2) + ((1-4)**2)) ** (1/2)
print(f"1. ‡∏û‡∏∑‡πâ‡∏ô‡∏ó‡∏µ‡πà‡∏ß‡∏á‡∏Å‡∏•‡∏° = {A:.2f}")
print(f"2. ‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå‡∏ô‡∏¥‡∏û‡∏à‡∏ô‡πå = {B:.2f}")
print(f"3. ‡∏£‡∏≤‡∏Å‡∏ó‡∏µ‡πà 3 ‡∏Ç‡∏≠‡∏á 64 = {C:.2f}")
print(f"4. ‡∏£‡∏∞‡∏¢‡∏∞‡∏ó‡∏≤‡∏á = {D:.2f}")
# ‡∏û‡∏¥‡∏°‡∏û‡πå‡πÇ‡∏Ñ‡πâ‡∏î‡∏Ñ‡∏≥‡∏ï‡∏≠‡∏ö‡∏ó‡∏µ‡πà‡∏ô‡∏µ‡πàf
````

‡πÉ‡∏´‡πâ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡πÇ‡∏õ‡∏£‡πÅ‡∏Å‡∏£‡∏°‡πÄ‡∏û‡∏∑‡πà‡∏≠:

   - A: ‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô >= 80

```python
# ‡πÉ‡∏´‡πâ‡∏£‡∏±‡∏ö‡∏Ñ‡πà‡∏≤‡∏à‡∏≤‡∏Å‡∏ú‡∏π‡πâ‡πÉ‡∏ä‡πâ‡∏´‡∏£‡∏∑‡∏≠‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏Ñ‡πà‡∏≤‡πÄ‡∏≠‡∏á
```

**Expected Output (‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏î‡πâ‡∏≤‡∏ô‡∏ö‡∏ô):**

```
‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡∏ß‡∏¥‡∏ä‡∏≤‡∏ó‡∏µ‡πà 3: 80
```

````python
# ‡∏û‡∏¥‡∏°‡∏û‡πå‡πÇ‡∏Ñ‡πâ‡∏î‡∏Ñ‡∏≥‡∏ï‡∏≠‡∏ö‡∏ó‡∏µ‡πà‡∏ô‡∏µ‡πà

# ‡∏™‡∏≤‡∏°‡∏≤‡∏£‡∏ñ‡∏£‡∏±‡∏ö‡∏Ñ‡πà‡∏≤‡∏à‡∏≤‡∏Å‡∏ú‡∏π‡πâ‡πÉ‡∏ä‡πâ‡∏´‡∏£‡∏∑‡∏≠‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏Ñ‡πà‡∏≤‡πÄ‡∏≠‡∏á
score1, score2, score3 = 75, 65, 80

print(f"‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡∏ß‡∏¥‡∏ä‡∏≤‡∏ó‡∏µ‡πà 1: {score1}")

print(f"‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡∏ß‡∏¥‡∏ä‡∏≤‡∏ó‡∏µ‡πà 2: {score2}")

print(f"‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡∏ß‡∏¥‡∏ä‡∏≤‡∏ó‡∏µ‡πà 3: {score3}")

scores = [score1, score2, score3]
avg = (score1 + score2 + score3) / len(scores)
print(f"‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢: {avg:.2f}")

if avg >= 50 and min(scores)>= 40:
    print("‡∏ú‡∏•‡∏Å‡∏≤‡∏£‡∏™‡∏≠‡∏ö: ‡∏ú‡πà‡∏≤‡∏ô")
else:
    print("‡∏ú‡∏•‡∏Å‡∏≤‡∏£‡∏™‡∏≠‡∏ö: ‡πÑ‡∏°‡πà‡∏ú‡πà‡∏≤‡∏ô:")

if avg >= 80:
    print("‡πÄ‡∏Å‡∏£‡∏î = A")
elif avg >= 70:
    print("‡πÄ‡∏Å‡∏£‡∏î = B")
elif avg >= 60:
    print("‡πÄ‡∏Å‡∏£‡∏î = C")
elif avg >= 50:
    print("‡πÄ‡∏Å‡∏£‡∏î = D")
else:
    print("‡πÄ‡∏Å‡∏£‡∏î = F")
````

### ‡∏Ç‡πâ‡∏≠ 6: ‡∏Å‡∏≤‡∏£‡∏ß‡∏ô‡∏ã‡πâ‡∏≥‡πÅ‡∏•‡∏∞‡∏Å‡∏≤‡∏£‡∏Å‡∏£‡∏≠‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• (Loop with Filtering)

‡πÉ‡∏´‡πâ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡πÇ‡∏õ‡∏£‡πÅ‡∏Å‡∏£‡∏°‡∏ó‡∏µ‡πà‡πÅ‡∏™‡∏î‡∏á‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç 1 ‡∏ñ‡∏∂‡∏á 50 ‡πÇ‡∏î‡∏¢‡πÉ‡∏ä‡πâ loop ‡πÅ‡∏•‡∏∞‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£‡∏î‡∏±‡∏á‡∏ô‡∏µ‡πâ:
1. ‡∏ñ‡πâ‡∏≤‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡∏´‡∏≤‡∏£‡∏î‡πâ‡∏ß‡∏¢ 3 ‡∏•‡∏á‡∏ï‡∏±‡∏ß ‡πÉ‡∏´‡πâ‡∏Ç‡πâ‡∏≤‡∏°‡πÑ‡∏õ (‡πÉ‡∏ä‡πâ continue)
2. ‡∏ñ‡πâ‡∏≤‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡∏´‡∏≤‡∏£‡∏î‡πâ‡∏ß‡∏¢ 7 ‡∏•‡∏á‡∏ï‡∏±‡∏ß ‡πÉ‡∏´‡πâ‡πÅ‡∏™‡∏î‡∏á "LUCKY" ‡πÅ‡∏ó‡∏ô‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç
3. ‡∏Å‡∏£‡∏ì‡∏µ‡∏≠‡∏∑‡πà‡∏ô ‡πÉ‡∏´‡πâ‡πÅ‡∏™‡∏î‡∏á‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡∏ï‡∏≤‡∏°‡∏õ‡∏Å‡∏ï‡∏¥
4. ‡∏ñ‡πâ‡∏≤‡πÄ‡∏à‡∏≠‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡∏ó‡∏µ‡πà‡∏´‡∏≤‡∏£‡∏î‡πâ‡∏ß‡∏¢ 37 ‡∏•‡∏á‡∏ï‡∏±‡∏ß ‡πÉ‡∏´‡πâ‡∏´‡∏¢‡∏∏‡∏î‡∏Å‡∏≤‡∏£‡∏ß‡∏ô‡∏ã‡πâ‡∏≥ (‡πÉ‡∏ä‡πâ break)

**Expected Output (‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£‡πÅ‡∏£‡∏Å‡πÜ):**
```
1 2 4 5 LUCKY 8 10 11 13 LUCKY 16 17 19 20 22 23 25 26 LUCKY 29 31 32 34 35 ...
```

````python
# ‡∏û‡∏¥‡∏°‡∏û‡πå‡πÇ‡∏Ñ‡πâ‡∏î‡∏Ñ‡∏≥‡∏ï‡∏≠‡∏ö‡∏ó‡∏µ‡πà‡∏ô‡∏µ‡πà
for i in range(0,51):
    if i % 3 == 0:
        continue
    elif i % 7 == 0:
        print("LUCKY")
    elif i % 37 == 0:
        break
    else:
        print(i)
    
````

### ‡∏Ç‡πâ‡∏≠ 8: ‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏™‡∏ñ‡∏¥‡∏ï‡∏¥ (Function for Statistics)

‡πÉ‡∏´‡πâ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô `calculate_stats(numbers)` ‡∏ó‡∏µ‡πà:
1. ‡∏£‡∏±‡∏ö list ‡∏Ç‡∏≠‡∏á‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡πÄ‡∏õ‡πá‡∏ô parameter
2. ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡πÅ‡∏•‡∏∞ return dictionary ‡∏ó‡∏µ‡πà‡∏°‡∏µ keys ‡∏ï‡πà‡∏≠‡πÑ‡∏õ‡∏ô‡∏µ‡πâ:
   - `min`: ‡∏Ñ‡πà‡∏≤‡∏ó‡∏µ‡πà‡∏ô‡πâ‡∏≠‡∏¢‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î
   - `max`: ‡∏Ñ‡πà‡∏≤‡∏ó‡∏µ‡πà‡∏°‡∏≤‡∏Å‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î
   - `sum`: ‡∏ú‡∏•‡∏£‡∏ß‡∏°‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î
   - `count`: ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç
   - `average`: ‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢ (rounded to 2 decimal places)

**‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏Å‡∏≤‡∏£‡πÉ‡∏ä‡πâ:**
```python
scores = [78, 85, 92, 64, 55, 88, 71, 95, 42, 80]
result = calculate_stats(scores)
print(result)
```

**Expected Output:**
```
{'min': 42, 'max': 95, 'sum': 750, 'count': 10, 'average': 75.0}
```

````python
# ‡∏û‡∏¥‡∏°‡∏û‡πå‡πÇ‡∏Ñ‡πâ‡∏î‡∏Ñ‡∏≥‡∏ï‡∏≠‡∏ö‡∏ó‡∏µ‡πà‡∏ô‡∏µ‡πà
dic = {}
# ‡πÉ‡∏´‡πâ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô calculate_stats ‡πÅ‡∏•‡πâ‡∏ß‡πÄ‡∏£‡∏µ‡∏¢‡∏Å‡πÉ‡∏ä‡πâ
scores = [78, 85, 92, 64, 55, 88, 71, 95, 42, 80]
def calculate_stats(numbers):
    result = {
        "min": min(numbers),
        "max": max(numbers),
        "sum": sum(numbers),
        "count": len(numbers),
        "average": sum(numbers) / len(numbers)
    }
    return result
result = calculate_stats(scores)
print(result)
````

### ‡∏Ç‡πâ‡∏≠ 9: ‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏£‡∏´‡∏±‡∏™‡∏ú‡πà‡∏≤‡∏ô (Password Validation Function)

‡πÉ‡∏´‡πâ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô `check_password(password)` ‡∏ó‡∏µ‡πà‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏Ñ‡∏ß‡∏≤‡∏°‡∏õ‡∏•‡∏≠‡∏î‡∏†‡∏±‡∏¢‡∏Ç‡∏≠‡∏á‡∏£‡∏´‡∏±‡∏™‡∏ú‡πà‡∏≤‡∏ô:

**‡πÄ‡∏á‡∏∑‡πà‡∏≠‡∏ô‡πÑ‡∏Ç‡∏Å‡∏≤‡∏£‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö:**
1. ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏¢‡∏≤‡∏ß‡∏ï‡πâ‡∏≠‡∏á‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ô‡πâ‡∏≠‡∏¢ 8 ‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£
2. ‡∏ï‡πâ‡∏≠‡∏á‡∏°‡∏µ‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç (0-9) ‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ô‡πâ‡∏≠‡∏¢ 1 ‡∏ï‡∏±‡∏ß
3. ‡∏ï‡πâ‡∏≠‡∏á‡∏°‡∏µ‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£‡∏û‡∏¥‡∏°‡∏û‡πå‡πÉ‡∏´‡∏ç‡πà (A-Z) ‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ô‡πâ‡∏≠‡∏¢ 1 ‡∏ï‡∏±‡∏ß

**‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏à‡∏∞ return:**
- `True` ‡∏ñ‡πâ‡∏≤ password ‡∏õ‡∏•‡∏≠‡∏î‡∏†‡∏±‡∏¢
- `False` ‡∏ñ‡πâ‡∏≤ password ‡πÑ‡∏°‡πà‡∏õ‡∏•‡∏≠‡∏î‡∏†‡∏±‡∏¢ ‡∏û‡∏£‡πâ‡∏≠‡∏°‡∏Å‡∏±‡∏ö‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ö‡∏≠‡∏Å‡πÄ‡∏´‡∏ï‡∏∏‡∏ú‡∏•

**‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏Å‡∏≤‡∏£‡πÉ‡∏ä‡πâ:**
```python
check_password("Hello123")     # True, "‡∏£‡∏´‡∏±‡∏™‡∏ú‡πà‡∏≤‡∏ô‡∏õ‡∏•‡∏≠‡∏î‡∏†‡∏±‡∏¢"
check_password("hello")         # False
check_password("Hello12")       # False
check_password("HELLO123")      # True
```

````python
# ‡∏û‡∏¥‡∏°‡∏û‡πå‡πÇ‡∏Ñ‡πâ‡∏î‡∏Ñ‡∏≥‡∏ï‡∏≠‡∏ö‡∏ó‡∏µ‡πà‡∏ô‡∏µ‡πà
# ‡πÉ‡∏´‡πâ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô  check_password ‡πÅ‡∏•‡πâ‡∏ß‡∏ó‡∏î‡∏™‡∏≠‡∏ö‡∏î‡πâ‡∏ß‡∏¢‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ï‡πà‡∏≤‡∏á‡πÜ
def check_password(pwd):
    if len(pwd) < 8:
        return False, "‡∏£‡∏´‡∏±‡∏™‡∏ú‡πà‡∏≤‡∏ô‡∏ï‡πâ‡∏≠‡∏á‡∏°‡∏µ‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ô‡πâ‡∏≠‡∏¢ 8 ‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£"
    if not any(char.isdigit() for char in pwd):
        return False, "‡∏ï‡πâ‡∏≠‡∏á‡∏°‡∏µ‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ô‡πâ‡∏≠‡∏¢ 1 ‡∏ï‡∏±‡∏ß"
    if not any(char.isupper() for char in pwd):
        return False, "‡∏ï‡πâ‡∏≠‡∏á‡∏°‡∏µ‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£‡∏û‡∏¥‡∏°‡∏û‡πå‡πÉ‡∏´‡∏ç‡πà‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ô‡πâ‡∏≠‡∏¢ 1 ‡∏ï‡∏±‡∏ß"
    return True,"‡∏£‡∏´‡∏±‡∏™‡∏ú‡πà‡∏≤‡∏ô‡∏õ‡∏•‡∏≠‡∏î‡∏†‡∏±‡∏¢"
test_passwords = ["Hello123", "hello", "Hello12", "HELLO123"]
for p in test_passwords:
    print(check_password(p))
    
````

### ‡∏Ç‡πâ‡∏≠ 10: ‡πÇ‡∏õ‡∏£‡πÅ‡∏Å‡∏£‡∏°‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£‡∏£‡πâ‡∏≤‡∏ô‡∏Ñ‡πâ‡∏≤ (Inventory Management System)

‡πÉ‡∏´‡πâ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡πÇ‡∏õ‡∏£‡πÅ‡∏Å‡∏£‡∏°‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£‡∏Ñ‡∏•‡∏±‡∏á‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡∏ó‡∏µ‡πà‡∏°‡∏µ dictionary ‡πÄ‡∏Å‡πá‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤ ‡πÅ‡∏•‡∏∞‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏ï‡πà‡∏≤‡∏á‡πÜ ‡∏î‡∏±‡∏á‡∏ô‡∏µ‡πâ:

**‡πÇ‡∏Ñ‡∏£‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•:**
- Dictionary ‡∏ä‡∏∑‡πà‡∏≠ `inventory` ‡∏°‡∏µ structure: `{"‡∏ä‡∏∑‡πà‡∏≠‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤": {"price": ‡∏£‡∏≤‡∏Ñ‡∏≤, "quantity": ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô}}`

**‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏ó‡∏µ‡πà‡∏ï‡πâ‡∏≠‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á:**

3. `show_inventory()` - ‡πÅ‡∏™‡∏î‡∏á‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î ‡∏û‡∏£‡πâ‡∏≠‡∏° ‡∏£‡∏≤‡∏Ñ‡∏≤ ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô ‡∏°‡∏π‡∏•‡∏Ñ‡πà‡∏≤
4. `total_value()` - ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏°‡∏π‡∏•‡∏Ñ‡πà‡∏≤‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î (‡∏£‡∏≤‡∏Ñ‡∏≤ √ó ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô)

**‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏Å‡∏≤‡∏£‡πÉ‡∏ä‡πâ:**

```python
print(f"‡∏°‡∏π‡∏•‡∏Ñ‡πà‡∏≤‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î: {total_value()} ‡∏ö‡∏≤‡∏ó")
```

````python
# ‡∏û‡∏¥‡∏°‡∏û‡πå‡πÇ‡∏Ñ‡πâ‡∏î‡∏Ñ‡∏≥‡∏ï‡∏≠‡∏ö‡∏ó‡∏µ‡πà‡∏ô‡∏µ‡πà

inventory = {}
# ‡πÉ‡∏´‡πâ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô add_product, sell_product, show_inventory, total_value
inventory = {}

def add_product(name, price, quantity):
    inventory[name] = {
        "price": price,
        "quantity": quantity
    }

def sell_product(name, quantity):
    if name in inventory and inventory[name]["quantity"] >= quantity:
        inventory[name]["quantity"] -= quantity

def show_inventory():
    print("===== ‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤ =====")
    i = 1
    for name, data in inventory.items():
        price = data["price"]
        quantity = data["quantity"]
        value = price * quantity

        print(f"{i}. {name}")
        print(f"‡∏£‡∏≤‡∏Ñ‡∏≤: {price} ‡∏ö‡∏≤‡∏ó/‡∏ä‡∏¥‡πâ‡∏ô")
        print(f"‡∏Ñ‡∏á‡πÄ‡∏´‡∏•‡∏∑‡∏≠: {quantity} ‡∏ä‡∏¥‡πâ‡∏ô")
        print(f"‡∏°‡∏π‡∏•‡∏Ñ‡πà‡∏≤: {value} ‡∏ö‡∏≤‡∏ó")
        print()
        i += 1

def total_value():
    total = 0
    for data in inventory.values():
        total += data["price"] * data["quantity"]
    return total

# ‡πÅ‡∏•‡πâ‡∏ß‡∏ó‡∏î‡∏™‡∏≠‡∏ö‡∏î‡πâ‡∏ß‡∏¢‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏î‡πâ‡∏≤‡∏ô‡∏ö‡∏ô
# add_product("‡∏ô‡∏°", 25, 10)
add_product("‡∏ô‡∏°", 25, 10)
# add_product("‡∏Ç‡∏ô‡∏°‡∏õ‡∏±‡∏á", 15, 20)
add_product("‡∏Ç‡∏ô‡∏°‡∏õ‡∏±‡∏á", 15, 20)
# add_product("‡πÑ‡∏Ç‡πà", 30, 5)
add_product("‡πÑ‡∏Ç‡πà", 30, 5)
# sell_product("‡∏ô‡∏°", 3)‡∏´
sell_product("‡∏ô‡∏°", 3)
# show_inventory()
show_inventory()
# print(f"‡∏°‡∏π‡∏•‡∏Ñ‡πà‡∏≤‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î: {total_value()} ‡∏ö‡∏≤‡∏ó")
print(f"‡∏°‡∏π‡∏•‡∏Ñ‡πà‡∏≤‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î: {total_value()} ‡∏ö‡∏≤‡∏ó")
````
